```python
%matplotlib inline
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

base_path = "C:/Users/User/Capstone/detail/detail/dialog"
save_dir = "C:/Users/User/Capstone/figures"   # folder to save figures
os.makedirs(save_dir, exist_ok=True)

for fold in range(5):  # fold=0 to fold=4
    fold_path = os.path.join(base_path, f"fold={fold}")
    print(f"\n===== Now loading fold {fold} =====")
    
    df = pd.read_parquet(fold_path)
    print(df.shape)
    print(df.head(10))

    # ===== Basic cleaning =====
    df["event_time"] = pd.to_datetime(df["event_time"])
    print("Missing values:\n", df.isna().sum())

    # ===== Descriptive statistics =====
    n_clients = df["client_id"].nunique()
    print("Number of unique clients:", n_clients)

    # ===== Temporal patterns =====
    daily_events = df.groupby(df["event_time"].dt.date).size()
    ax = daily_events.plot(figsize=(12, 5), title=f"Daily Events (Fold {fold})")
    ax.set_xlabel("Date"); ax.set_ylabel("Event Count")
    plt.savefig(os.path.join(save_dir, f"daily_events_fold{fold}.png"), dpi=300, bbox_inches="tight")
    plt.show()

    monthly_events = df.groupby(df["event_time"].dt.to_period("M")).size()
    ax = monthly_events.plot(figsize=(12, 5), title=f"Monthly Events (Fold {fold})")
    ax.set_xlabel("Month"); ax.set_ylabel("Event Count")
    plt.savefig(os.path.join(save_dir, f"monthly_events_fold{fold}.png"), dpi=300, bbox_inches="tight")
    plt.show()

    # ===== Client-level analysis =====
    events_per_client = df.groupby("client_id").size()
    ax = events_per_client.hist(bins=50)
    plt.title(f"Distribution of Events per Client (Fold {fold})")
    plt.xlabel("Number of Events"); plt.ylabel("Number of Clients")
    plt.savefig(os.path.join(save_dir, f"events_per_client_fold{fold}.png"), dpi=300, bbox_inches="tight")
    plt.show()

    print("Top 10 most active clients:\n", events_per_client.sort_values(ascending=False).head(10))

    # ===== Embedding clustering =====
    print("Clustering embeddings...")

    embeddings = np.vstack(df["embedding"].values)  # shape: (n_samples, n_dims)

    n_clusters = 10
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    df["cluster"] = kmeans.fit_predict(embeddings)

    print("Cluster sizes:\n", df["cluster"].value_counts())

    # ===== Visualize clusters (2D PCA) =====
    print("Visualizing clusters using PCA...")
    pca = PCA(n_components=2)
    emb_2d = pca.fit_transform(embeddings)

    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(emb_2d[:, 0], emb_2d[:, 1], c=df["cluster"], cmap="tab10", alpha=0.6)
    plt.title(f"Embedding Clusters (Fold {fold}) - PCA 2D")
    plt.xlabel("PCA-1"); plt.ylabel("PCA-2")
    plt.colorbar(scatter, label="Cluster ID")
    plt.savefig(os.path.join(save_dir, f"embedding_clusters_pca_fold{fold}.png"), dpi=300, bbox_inches="tight")
    plt.show()
```
